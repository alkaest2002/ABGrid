```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
title = "prova"

fig, ax = plt.subplots(figsize=(8,8))
ax.axis("off")
G.add_edges_from([(1,2),(1,3),(1,5),(2,3),(3,4),(4,5)])
loc = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

nx.draw_networkx_nodes(G, loc, node_color="#fff", node_size=1000, edgecolors="k", ax=ax)
nx.draw_networkx_edges(G, loc)
nx.draw_networkx_labels(G, loc, font_size=22)

plt.savefig(f"{title}.png",  bbox_inches='tight', transparent=True, pad_inches=0)
```
